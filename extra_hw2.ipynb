```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: ((-(x**2)) / 100) + (10 * math.sin(x))
```

```python
def derivative(f, x1):
    dx = 0.0001
    x2 = x1 + dx
    y1 = f(x1)
    y2 = f(x2)
    dy = y2 - y1
    return dy / dx
```

```python
def gradient_descent(f, x):
    start = x
    for i in range(1000):
        der = derivative(f, x)
        if der < 0:
            x -= 0.1
        else:
            x += 0.1
            
    return 'start x is {}, final x is {}, final y is {}'.format(start,x,f(x))
```

```python
x1 = np.arange(-20, 20, 0.0001)   
f2 = np.vectorize(f)                                
plt.plot(x, f2(x))
plt.show()   
```

```python
for i in range(5):
    x = random.randint(-20, 20)
    print(gradient_descent(f, x))
```
