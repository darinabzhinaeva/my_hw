```python
def F(a, b, c):
    return 3 * a**30 + b + 5 * c**2
```

```python
def derivative_a(f, a, b, c):
    da = 0.0001
    f1 = f(a, b, c)
    f2 = f(a + da, b, c)
    df = f2 - f1
    return df / da



def derivative_b(f, a, b, c):
    db = 0.0001
    f1 = f(a, b, c)
    f2 = f(a, b + db, c)
    df = f2 - f1
    return df / db



def derivative_c(f, a, b, c):
    dc = 0.0001
    f1 = f(a, b, c)
    f2 = f(a, b, c + dc)
    df = f2 - f1
    return df / dc
```

```python
print(derivative_a(F, 1, 2, 3))
print(derivative_b(F, 1, 2, 3))
print(derivative_c(F, 1, 2, 3))
```

```python
derivative_a(F, 10, 2, 3)
```

```python
for k in range(-1000, 1000):
    print(k, derivative_b(F, 1, k, 3))# we will never have derivative -7 in this linear function
```

```python
derivative_c(F, 1, 2, 1.5)
```
