```python
from matplotlib.pyplot import plot
import numpy as np
import math
```

```python
f1 = lambda x: math.sin(x)
f2 = lambda y: math.log(y)
```

```python
def derivative(f, x):
    dx = 0.0001
    f1 = f(x)
    f2 = f(x + dx)
    df = f2 - f1
    return df / dx
```

```python
l1 = []
for i in np.arange(0.001, 10, 0.01):
    l1.append(derivative(f1, i))
```

```python
l2 = []
for k in np.arange(0.001, 10, 0.01):
    l2.append(derivative(f2, k))
```

```python
plot(range(len(l1)), l1)
```

```python
plot(range(len(l2)), l2)
```

```python
f3 = lambda x, y: (math.cos(x))**2 - (y-3)**2
```

```python
def plot_max(f, left, right, step):
    points = []
    for x in np.arange(left, right, step):
        for y in np.arange(left, right, step):
            points.append(f3(x, y))
    return points
```

```python
m = plot_max(f3, -10, 10, 0.01)
print(sorted(m)[-10:])
```
