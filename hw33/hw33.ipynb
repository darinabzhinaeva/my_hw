```python
import pandas as pd
actors = pd.DataFrame(columns=['name'])
actors.index = actors.index.rename('actor_id')
actors.loc[0] = ['Johny Depp']
actors.loc[1] = ['Leonardo DiCaprio']
actors.loc[2] = ['Natalie Portman']
actors.loc[3] = ['Will Smith']
actors.loc[4] = ['Emma Watson']
actors.head()
```

```python
movies = pd.DataFrame(columns=['name'])
movies.index = movies.index.rename('movie_id')
movies.loc[0] = ['Pirates of the Caribbean']
movies.loc[1] = ['Titanic']
movies.loc[2] = ['Thor']
movies.loc[3] = ['Men in Black ']
movies.loc[4] = ['Little Women']
movies.head()
```

```python
companies = pd.DataFrame(columns=['name'])
companies.index = companies.index.rename('companie_id')
companies.loc[0] = ['Overbrook Entertainment']
companies.loc[1] = ['Infinitum Nihil']
companies.loc[2] = ['Warner Bros']
companies.loc[3] = ['Handsomecharlie Films']
companies.loc[4] = ['Appian Way Productions']
companies.head()
```

```python
tmp1 = pd.merge(actors, movies, left_index=True, right_on='movie_id')
tmp1.head()
```

```python
tmp2 = pd.merge(movies, companies, left_index=True, right_on='companie_id')
tmp2.head()
```

```python
tmp3 = pd.merge(tmp1, companies, left_index=True, right_on='companie_id')
tmp3.head()
```
