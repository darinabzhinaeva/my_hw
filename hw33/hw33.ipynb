```python
import pandas as pd
actors = pd.DataFrame(columns=['name'])
actors.index = actors.index.rename('actor_id')
actors.loc[0] = ['Johny Depp']
actors.loc[1] = ['Leonardo DiCaprio']
actors.loc[2] = ['Natalie Portman']
actors.loc[3] = ['Will Smith']
actors.loc[4] = ['Emma Watson']
actors.head()
```

```python
movies = pd.DataFrame(columns=['name', 'company_id'])
movies.index = movies.index.rename('movie_id')
movies.loc[0] = ['Pirates of the Caribbean', 0]
movies.loc[1] = ['Titanic', 0]
movies.loc[2] = ['Thor', 1]
movies.loc[3] = ['Men in Black ', 1]
movies.loc[4] = ['Little Women', 2]
movies.head()
```

```python
companies = pd.DataFrame(columns=['name'])
companies.index = companies.index.rename('company_id')
companies.loc[0] = ['Overbrook Entertainment']
companies.loc[1] = ['Infinitum Nihil']
companies.loc[2] = ['Warner Bros']
companies.loc[3] = ['Handsomecharlie Films']
companies.loc[4] = ['Appian Way Productions']
companies.head()
```

```python
actors_movies = pd.DataFrame(columns=['actor_id', 'movie_id'])
actors_movies.loc[0] = [0, 1]
actors_movies.loc[1] = [0, 2]
actors_movies.loc[2] = [1, 0]
actors_movies.loc[3] = [3, 2]
actors_movies.loc[4] = [4, 3]
actors_movies.loc[5] = [1, 3]
actors_movies.loc[6] = [2, 2]
actors_movies.loc[7] = [0, 1]
actors_movies.loc[8] = [2, 0]
actors_movies.loc[9] = [4, 2]
actors_movies.loc[10] = [0, 3]
actors_movies
```

```python
pd.merge(movies, actors_movies, left_index=True, right_on='movie_id')
```

```python
pd.merge(companies, movies, left_index=True, right_on='company_id')
```

```python
tmp = pd.merge(pd.merge(pd.merge(movies, actors_movies, on='movie_id'), actors, on='actor_id'), companies, on='company_id')
tmp[['company_id', 'movie_id', 'actor_id', 'name_x', 'name_y', 'name']]
```
