```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def derivative(f, x1):
    dx = 0.0001
    x2 = x1 + dx
    y1 = f(x1)
    y2 = f(x2)
    dy = y2 - y1
    return dy / dx
```

```python
f1 = lambda x: (x / 10) ** 3 * math.sin(2 * x)    # this 
```

```python
x = np.arange(-50, 50, 0.0001)   
f2 = np.vectorize(f1)                                
plt.plot(x, f2(x))
plt.show()   
```

```python
def gradient_descent(f, x):
    start = x
    for i in range(1000):
        der = derivative(f, x)
        if der < 0:
            x += 0.1
        else:
            x -= 0.1
            
    return 'start x is {}, final x is {}, final y is {}'.format(start,x,f(x))
```

```python
for i in range(3):
    x = random.randint(-50, 50)
    print(gradient_descent(f1, x))
```

```python
f3 = lambda x: (x**4) - (x**3) + (x**-1)
for i in range(3):
    x = random.randint(-50, 50)
    print(gradient_descent(f3, x))
```
