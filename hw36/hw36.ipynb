```python
import pandas as pd
qid_query = pd.read_csv("d_task/qid_query.tsv", sep="\t", names=["qid", "query"])
qid_url_rating = pd.read_csv("d_task/qid_url_rating.tsv", sep="\t", names=["qid", "url", "rating"])
hostid_url = pd.read_csv("d_task/hostid_url.tsv", sep="\t", names=["hostid", "url"])
qid_query.head()
```

```python
qid_url_rating_hostid = pd.merge(qid_url_rating, hostid_url, on="url")
qid_url_rating_hostid.head()
```

```python
qid_url_rating_hostid['len'] = qid_url_rating_hostid['url'].map(len)
qid_url_rating_hostid.groupby('qid').min()
```

```python
def calculatePLook(pRel):
    pBreak = 0.15
    pLook = [1]
    for rel in pRel[:10]:
        pLook.append(pLook[-1] * (1 - rel) * (1 - pBreak))
    return pLook

def pFound(pRel):
    pLook = calculatePLook(pRel)
    return sum([pLook[i] * pRel[i] for i in range(10)]) 
```

```python
for _id in qid_query.qid.unique():
    tmp = qid_url_rating_hostid[qid_url_rating_hostid.qid == _id].groupby('hostid').max()
    pRel = tmp.sort_values('rating', ascending=False)['rating'][:10].values
    print(_id, pFound(pRel))
```
