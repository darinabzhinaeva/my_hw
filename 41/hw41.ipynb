```python
import pandas as pd
all_ratings = pd.read_csv('./u.data', sep='\t', header=None)
all_ratings.columns = ['user_id', 'item_id', 'rating', 'timestamp']
all_ratings.user_id = all_ratings.user_id - 1
all_ratings.item_id = all_ratings.item_id - 1
all_ratings
```

```python
all_ratings.groupby('item_id').min()['user_id']     #Min of user_id for every item_id
```

```python
all_ratings.groupby('item_id').mean()['timestamp']   #Average of timestamp for every item_id
```

```python
all_ratings.groupby('user_id').mean()['timestamp']    #Average of timestamp for every user_id
```

```python
import numpy as np
matrix = np.zeros(shape=(100, 100), dtype='int')
matrix = matrix.tolist()
for row in range(len(matrix)):
    for val in range(len(matrix[0])):
        if (row + val) % 5 == 0:
            matrix[row][val] = 1
for r in matrix:
    print(r)
```

```python
coo = []
for a in range(len(matrix)):
    for b in range(len(matrix[0])):
        if matrix[a][b] == 1:
            coo.append([])
            coo[-1].append(a)
            coo[-1].append(b)
            coo[-1].append(1)
        
for uu in coo:
    print(uu)
```
