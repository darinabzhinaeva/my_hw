```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def plot_min(f, left, right, step):
    minX = left
    minY = f(left)
    for i in np.arange(left, right, step):
        if minY > f(i):
            minY = f(i)
            minX = i
    return minX, minY
```

```python
def plot_max(f, left, right, step):
    maxX = left
    maxY = f(left)
    for i in np.arange(left, right, step):
        if maxY < f(i):
            maxY = f(i)
            maxX = i
    return maxX, maxY
```

```python
fmin = lambda x: (x-0.1)*(x-0.1)/10 + math.sin(3*x)
left = -10
right = 10
step = 0.1
minnX = left
minnY = fmin(left)
res_step = 0
for k in range(100000):
    t = plot_min(fmin, left, right, step)
    if t[1] < minnY:
        minnX = t[0]
        minnY = t[1]
        res_step = step
    step += 0.000001
        
print(minnX, minnY, res_step)
```

```python
fmax = lambda x: math.sin(x) + math.sqrt(abs(x-5))
left = -20
right = 15
step = 0.001
maxxX = left
maxxY = fmax(left)
res_step = 0
for k in range(100000):
    t = plot_max(fmax, left, right, step)
    if t[1] > maxxY:
        maxxX = t[0]
        maxxY = t[1]
        res_step = step
    step += 0.000001
        
print(maxxX, maxxY, res_step)
```
