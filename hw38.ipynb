```python
import pandas as pd
```

```python
all_ratings = pd.read_csv('light_fm_F_task/u.data', sep='\t', header=None)
all_ratings.columns = ['user_id', 'item_id', 'rating', 'timestamp']
all_ratings.user_id = all_ratings.user_id - 1
all_ratings.item_id = all_ratings.item_id - 1
all_ratings
```

```python
items = pd.read_csv('light_fm_F_task/u.item', sep="|", header=None)
items.columns = ["item_id", "movie_title", "release_date", "video_release_date",
              "IMDb_URL", "unknown", "Action", "Adventure","Animation",
              "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy",
              "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi",
              "Thriller", "War", "Western"]
items.item_id = items.item_id - 1
items = items.set_index('item_id')
items = items[['movie_title']]
items
```

```python
items_rating = pd.merge(all_ratings, items, on='item_id')
items_rating
```

```python
sums = {}
for _id in all_ratings.item_id.unique():
    sums[str(_id)] = items_rating[items_rating.item_id == _id]['rating'].sum()
max_key = max(sums, key=sums.get)  
min_key = min(sums, key=sums.get)  
print('Film with maximum count of ratings is  ', list(items_rating[items_rating.item_id == int(max_key)].movie_title)[0]) 
print('Film with minimum count of ratings is  ', list(items_rating[items_rating.item_id == int(min_key)].movie_title)[0]) 
```

```python
means = {}
for _id in all_ratings.item_id.unique():
    means[str(_id)] = items_rating[items_rating.item_id == _id]['rating'].mean()
means = sorted(means.items(), key=lambda x: x[1], reverse=True)
for m in means[:10]:
    print(list(items_rating[items_rating.item_id == int(m[0])].movie_title)[0], 'rating is ', m[1])
```
