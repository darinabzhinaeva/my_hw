```python
def derivative(f, x1):
    dx = 0.0001
    x2 = x1 + dx
    y1 = f(x1)
    y2 = f(x2)
    dy = y2 - y1
    return dy / dx
```

```python
import math
import random
def gradient_descent(f, learning_rate=0.001, iteration_count=100, start_point=None):
    if start_point is None:
        r = random.randint(-10, 10)
    else:
        r = start_point
    print(r)
    for i in range(iteration_count):
        d = derivative(f, r)
        r -= learning_rate * d
        print(r)
    return r
```
